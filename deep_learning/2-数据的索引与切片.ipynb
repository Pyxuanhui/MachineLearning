## 2.1.3 广播机制

1. 广播机制的工作方式：
    - 通过适当复制元素来扩展一个或两个数组，以便在转换之后，两个张量具有相同的形状
    - 对生成的数组执行按元素操作
    

```python
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
# a 为 3×1 矩阵
# b 为 1×2 矩阵

print(a)
print(b)

# 广播机制，将两个形状不同的张量，扩展为更大的矩阵
# 扩展为 3 × 2 矩阵
print(a + b)
```
