# KNN回归算法

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor

# 生成示例数据
X = np.random.rand(100, 1) * 10 # 特征
y = np.sin(X).ravel() + 0.1 * np.random.randn(100) # 添加一些噪声

# 拆分为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) # 70%训练，30%测试

# 创建KNN回归模型
knn_reg = KNeighborsRegressor(n_neighbors=5) # 使用5个邻居

# 训练模型
knn_reg.fit(X_train, y_train) # 训练KNN回归模型

# 在测试集上进行预测
y_pred = knn_reg.predict(X_test) # 预测测试集结果

# 可视化回归结果
plt.scatter(X_test, y_test, color='red', label='True Values') # 真实值
plt.scatter(X_test, y_pred, color='blue', label='Predicted Values') # 预测值
plt.title("KNN Regression") # 标题
plt.xlabel("Feature") # 横坐标
plt.ylabel("Target") # 纵坐标
plt.legend() # 图例
plt.show() # 显示图形
```
